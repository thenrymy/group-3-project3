```python
# Python SQL toolkit and Object Relational Mapper
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python
#Connect to PostgreSQL using psycopg2 driver.  Need to install using 'pip install sqlalchemy psycopg2-binary'
engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/real-estate_db")
conn = engine.connect()
```

```python
#Using the connection, read a table and store in a df
perth_properties_data = pd.read_sql("SELECT * FROM perth_properties", conn)
```

```python
# created a query that joins perth_properties and school_list to obtain a df of the top 50 schools by rank.

query = """
    SELECT DISTINCT nearest_school AS School, nearest_school_rank AS rank, school_list.latitude, school_list.longitude
    FROM perth_properties
    INNER JOIN school_list ON perth_properties.nearest_school = school_list.school_name
    WHERE nearest_school_rank IS NOT NULL
    ORDER BY nearest_school_rank
    LIMIT 50
"""

school_rank = pd.read_sql(query, conn)
school_rank
```
